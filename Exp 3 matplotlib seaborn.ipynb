```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/Rahul Soni/Downloads/aids-codes/Heart.csv")
df
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
display(df.dtypes)
```

```python
plt.scatter(df['Age'], df['Cholesterol'])
plt.title("Age vs Cholesterol")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.plot()
```

```python
df['Heart Disease'].value_counts().plot(kind='bar')
plt.xlabel('Heart Disease')
plt.ylabel("Status")
plt.title("Presence of heart disease")
plt.plot()
```

```python
sex_heart_disease = df.groupby(['Sex', 'Heart Disease']).size().unstack()
sex_heart_disease
```

```python
sex_heart_disease.plot(kind='bar', stacked="True")
plt.title("Gender vs Heart Disease")
plt.xlabel("Gender (0=Female, 1=Male)")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

```python
import seaborn as sns
```

```python
sns.histplot(df['Cholesterol'], kde=True)
plt.title("Cholesterol Distribution")
plt.xlabel("Choleterol")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x='Heart Disease', y='Age', data=df)

plt.title("Heart Disease vs Age")
plt.xlabel("Heart Disease")
plt.ylabel("Age")
plt.plot()
```
