```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
df = pd.DataFrame(digits.data)
df['label'] = digits.target

print("Data Loaded successfully")
print(df.head())
```

```python
df.to_csv("mnist_sample_dataa.csv", index = False)
print("Data Saved successfully as mnist_sample-dataa.csv")
```

```python
data = pd.read_csv('mnist_sample_dataa.csv')
print(f'CSV Loaded. Shape: {data.shape}')
```

```python
X = data.drop('label', axis = 1)
Y = data['label']
```

```python
X = X / X.max().max()
print("Data normalized between 0 & 1")
```

```python
X_np = X.to_numpy()
Y_np = Y.to_numpy()
```

```python
print(f'Feature Array shape: {X_np.shape}')
print(f'Target Array shape: {Y_np.shape[0]}')
```

```python
sample = X_np[0].reshape(8, 8)
```

```python
plt.imshow(sample, cmap='gray')
plt.title(f'Label: {Y_np[0]}')
plt.axis("off")
plt.show()
```
