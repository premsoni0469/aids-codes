```python
# Cell 2: Imports, Distributions, and Conditional Probability Table
from pomegranate import *
```

```python
# Define initial probabilities for parent nodes
guest = DiscreteDistribution({'A': 1./3, 'B': 1./3, 'C': 1./3})
prize = DiscreteDistribution({'A': 1./3, 'B': 1./3, 'C': 1./3})
```

```python
# Define the conditional probability table for the 'monty' node
# This table is manually entered based on the rules of the game
# Format: [guest_choice, prize_location, monty_opens, probability]
monty = ConditionalProbabilityTable(
[['A', 'A', 'A', 0.0],
 ['A', 'A', 'B', 0.5],
 ['A', 'A', 'C', 0.5],
 ['A', 'B', 'A', 0.0],
 ['A', 'B', 'B', 0.0],
 ['A', 'B', 'C', 1.0],
 ['A', 'C', 'A', 0.0],
 ['A', 'C', 'B', 1.0],
 ['A', 'C', 'C', 0.0],
 ['B', 'A', 'A', 0.0],
 ['B', 'A', 'B', 0.0],
 ['B', 'A', 'C', 1.0],
 ['B', 'B', 'A', 0.5],
 ['B', 'B', 'B', 0.0],
 ['B', 'B', 'C', 0.5],
 ['B', 'C', 'A', 1.0],
 ['B', 'C', 'B', 0.0],
 ['B', 'C', 'C', 0.0],
 ['C', 'A', 'A', 0.0],
 ['C', 'A', 'B', 1.0],
 ['C', 'A', 'C', 0.0],
 ['C', 'B', 'A', 1.0],
 ['C', 'B', 'B', 0.0],
 ['C', 'B', 'C', 0.0],
 ['C', 'C', 'A', 0.5],
 ['C', 'C', 'B', 0.5],
 ['C', 'C', 'C', 0.0]], [guest, prize])
```

```python
# Cell 4: Model Definition
# Create the nodes for the network
s1 = Node(guest, name="guest")
s2 = Node(prize, name="prize")
s3 = Node(monty, name="monty")
```

```python
# Create the Bayesian Network
model = BayesianNetwork("Monty Hall Problem")
model.add_states(s1, s2, s3)
```

```python
# Cell 5: Model Edges and Bake
# Add the edges (dependencies)
# Monty's choice (s3) depends on the guest's choice (s1) and the prize's location (s2)
model.add_edge(s1, s3)
model.add_edge(s2, s3)

# Finalize the model structure
model.bake()
```

```python
# Cell 6: Probability Calculation
# Calculate the joint probability for specific scenarios
print(model.probability([['A','A','A'],
                       ['A','A','B'],
                       ['C','C','B']]))
```

```python
# Cell 7: Redundant Edge and Bake
# This part from the document is redundant but included for completeness
model.add_edge(s2, s3)
model.bake()
```

```python
# Cell 8: Probability Calculation
# Calculate probability again
print(model.probability([['A','A','A'],
                       ['A','A','B'],
                       ['C','C','B']]))
```

```python
# Cell 9: Prediction
# Predict Monty's move given guest and prize
print(model.predict([['A','B',None],
                     ['A','C',None],
                     ['C','B',None]]))
```

```python
# Cell 10: Prediction
# Predict different missing values
print(model.predict([['A','B',None],
                     ['A',None,'C'],
                     [None,'B','A']]))
```
