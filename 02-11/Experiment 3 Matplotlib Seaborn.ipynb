```python
# MATPLOTLIB

import matplotlib.pyplot as plt

plt.scatter(df['Age'], df['Cholesterol'])
plt.title("Age vs Cholesterol")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.show()
```

```python
# Histogram using MATPLOTLIB

import matplotlib.pyplot as plt

# Histogram of Heart Disease

df['Heart Disease'].value_counts().plot(kind='bar')
plt.title("Heart Disease")
plt.xlabel("Heart Disease")
plt.ylabel("Count")
plt.show()
```

```python
# BAR CHART USING MATPLOTLIB

import matplotlib.pyplot as plt
import pandas as pd

sex_heart_disease = df.groupby(['Sex', 'Heart Disease']).size().unstack()
sex_heart_disease.plot(kind='bar', stacked = True)
plt.title("Heart Disease by gender")
plt.xlabel("Gender (0=Female, 1=Male) ")
plt.ylabel("Count")

plt.xticks(rotation=0)
plt.legend(title="Heart Disease")

plt.show()
```

```python
# SEABORN

# HISTOGRAM OF CHOLESTEROL DISTRIBUTION

import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['Cholesterol'], kde=True)
plt.xlabel("Cholesterol")
plt.ylabel("Frequency")
plt.title("Distribution of Cholesterol")
plt.show()
```

```python
# SEABORN
import seaborn as sns
import matplotlib.pyplot as plt

# BOXPLOT OF CHOLESTEROL DISTRIBUTION

sns.boxplot(x='Heart Disease', y='Age', data=df)
plt.title("Age Distribution by Heart Disease")
plt.xlabel("Hear Disease")
plt.ylabel("Age")
plt.show()
```
